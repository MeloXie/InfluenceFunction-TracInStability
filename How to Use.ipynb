# Here we introduce what each file and folder mean

1. **SimpleOverallFrameWork.ipynb**
This is the real base sample code for our work. This provides a simple runnable sample to construct a synthetic dataset, train a model with respect to that, and acquire the influence estimation results for that. All our experiments are based on the code with different levels of modifications. Detailed explanations are given within the file. This file is given to have a quick idea of what we are doing in the experiment. **This file will demonstrate most things in detail, while the other specific experiment files will only explain those modified parts in detail. Recommend to understand the base code first and then use the other codes.**


2. **Folder - ClassImbalance**
This is the folder for the experiment on Class Imbalance. It contains two files:

        2.1. **ClassImbalanceExperiment.ipynb**
            This is the file to acquire the ranked influence list for the Class Imbalance Experiment. The main difference from the base code is that the number of samples in each class within the train and test set is set to be imbalanced. The level of imbalance could be regulated. After running the file, two ranked influence lists shall be stored within the folder. To make sure the label could be mapped with the sample's ID, a mapping file will be generated to store the label-ID pair. This and the ranked Influence lists shall be used within the later analysis code.

        2.2. **ClassImbalance-Lift.ipynb**
            This is the file to analyse the result of the Class Imbalance experiment. It will first give how many 1's samples (the minority group) appear on the top-k, and then produce the lift plot shown in our paper.

3. **Folder - NumberofSamples**
This is the folder for the experiment on the number of Samples. It contains two files:

        3.1. **NumberOfSamplesExperiment.ipynb**
            This is the file to acquire the ranked influence list for the number of Samples Experiment. It makes no difference with the base code. After running the file, two ranked influence lists shall be stored within the folder. To test on different numbers of samples, please change the code 'train_df = nested_train_dfs[7]' to modify the train set size. All the ranked Influence lists shall be used within the later analysis code.

        3.2. **ChunkAnalysis.ipynb**
            This is the file to analyse the result of the number of Samples experiment. It will read in all the ranked influence lists and then produce the normalised entropy plot shown in our paper.

4. **Folder - NumberofFeatures**
This is the folder for the experiment on the number of Features. It contains two files:

        4.1. **NumberOfFeaturesExperiment.ipynb**
            This is the file to acquire the ranked influence list for the Number of Features Experiment. The main difference from the base code is that noise features are added to the original train and test set. The default train and test set has 10 informative features, while a different number of noise features could be added to that. After running the file, two ranked influence lists shall be stored within the folder. To test on different numbers of Features, please change the code 'k= KS[8]' to modify the Feature size. All the ranked Influence lists shall be used within the later analysis code.

        4.2. **FeatureHeatmap.ipynb**
            This is the file to analyse the result of the number of Features experiment. It will read in all the ranked influence lists, compute the pair-wise weighted kendall tau and then produce the heatmap shown in our paper.

5. **Folder - OutliersToSensitivity**
This is the folder for the experiment on the Outliers to Sensitivity. It contains two files:

        5.1. **OutliersExperiment.ipynb**
            This is the file to acquire the ranked influence list for the Outliers to Sensitivity Experiment. The main difference from the base code is the loss value for each training point is saved in a file. Any other things are exactly the same as the base code. After running the file, two ranked influence lists shall be stored within the folder. All the ranked Influence lists and the loss files shall be used within the later analysis code.

        5.2. **InfluenceAndTrainLossFrequency.ipynb**
            This is the file to analyse the result of the Outliers to Sensitivity experiment. It will read in all the ranked influence lists and loss files, merge them and plot the Influence Score vs Sample Loss with a regression line fit inside.

6. **Folder - NeighborDistance**
This is the folder for the experiments on the Neighbour Distance. It contains four files with 2 experiments:

        6.1. **NeighborDistance Experiment**:
   
               6.1.1. **NeighborDistanceExperiment.ipynb**
                   This is the file to acquire the ranked influence list for the Neighbour Distance Experiment. It is exactly the same as the base code. After running the file, two ranked influence lists shall be stored within the folder. All the ranked Influence lists shall be used within the later analysis code.

               6.1.2. **RadiusExperimentAnalysis.ipynb**
            This is the file to analyse the result of the Neighbour Distance Experiment. It will read in one ranked influence list and corresponding experiment setting, check the relation of the influence difference with neighbours' distance, and then produce the corresponding plot shown in our paper.

       6.2. **Small Large Density Experiment**:
               6.2.1. **SmallLargeDensityExperiment**
                      This is the file to acquire the ranked influence list for the Small Large Density Experiment. The main difference from the base code is that one class is set to have two different clusters. One cluster is having strong local density, while the other has a large neighbour distance. To map those samples with the clusters, each pair of ID-cluster is stored inside a file. After running the file, two ranked influence lists shall be stored within the folder. The ranked Influence lists and mapping file shall be used within the later analysis code.

               6.1.2. **SmallLargeDensity-Comparison.ipynb**
            This is the file to analyse the result of the Small Large Density Experiment. It will read in one ranked influence list and corresponding mapping file, show the ranking position distribution of each sample, and then produce a corresponding plot based on the distribution, as shown in our paper.
